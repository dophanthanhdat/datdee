```python
import folium
from folium.plugins import HeatMap
import numpy as np

center = [10.76, 106.68]

m = folium.Map(location=center, zoom_start=12)

np.random.seed(42)

data = []

for _ in range(100):
    lat = 10.76 + np.random.normal(0, 0.005)
    lon = 106.68 + np.random.normal(0, 0.005)
    data.append([lat, lon])

for _ in range(60):
    lat = 10.75 + np.random.normal(0, 0.007)
    lon = 106.67 + np.random.normal(0, 0.007)
    data.append([lat, lon])

for _ in range(40):
    lat = 10.84 + np.random.normal(0, 0.01)
    lon = 106.83 + np.random.normal(0, 0.01)
    data.append([lat, lon])

HeatMap(data, radius=15, blur=10).add_to(m)

m
```
